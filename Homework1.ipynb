```python
x = np.array(Lotus_cars_Engine_columns_drop_duplicates)
x
```

```python
transpose_x = x.T
print(transpose_x)
xtx = transpose_x @ x
print(xtx)
inv_xtx = np.linalg.inv(xtx)
print(inv_xtx)
```
